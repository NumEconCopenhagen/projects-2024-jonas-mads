# Introduction: Analysis and Interactive Visualization

This project will conduct a detailed analysis of 4 economic indicators: Gross Domestic Product (GDP), Inflation (using the Consumer Price Index), Unemployment Rate, and the Federal Funds Rate. **GDP** measures the monetary value of good and services for a country. **Inflation** measures how much more expensive a set of goods and services has become over a certain period. **Unemployment Rate** measures the percentage of people in the labour force who are unemployed. **Federal Funds Rate** is the rate at which banks and other depository institutions lend money to each other. This project will consists of the following:

- **Descriptive Statistics**: We will start out with presenting a statistics for each of the indicators (mean, medians, and standard deviation) to get a better understanding of the data.
- **Correlation Analysis**: Thereafter, we will examine the correlation matrix to see the relationships between these indicators.
- **Visualization of Individual Variables**: We will thereafter plot each indicator separately to observe individual trends and variations over the period.
- **Interactive Visualization**: Lastly, we will focus on the relationship between the Inflation Rate and the Federal Funds Rate, by creating an interactive plot. This visualization will allow us to examine closely how these two variables have interacted over time. Using an interactive plot we are able to better examine certain periods to better see how the two indicators vary over time.

For the **GDP and Unemployment** we see a negative correlation here of -0.26, supporting Okun's law, which states that a rise in employment is often associated with a rise in GDP. Finally for the **Unemployment and Interest Rates** there is a strong negative correlation of -0.52, showing that higher unemployment typically coincides with lower interest rates. This strongly reflects the expansionary monetary policies during economic downturns.

## Federal Funds Rate

This graph displays the Federal Funds Rate over time. It indicates the interest rate at which depository institutions lend reserve balances to other depository institutions overnight.


```python
model.plot_interest_rate(model.data)
```

The above graph shows the US Federal Funds Rate from 1992 to 2024. Showing how the interest rate has changed throughout the years and influencing the economy. The fluctuations in the interest rate reflect the Federal Reserve's efforts to manage inflation, stimulate economic growth, and stabilize financial markets. Rate hikes aim to curb inflationary pressures during periods of economic expansion, while rate cuts aim to stimulate borrowing and spending during downturns. Notable outliers include the near-zero interest rate environment following the 2008 crisis and the 2020 COVID pandemic, signaling unconventional monetary policy measures to combat the aftermath of the recession. It is difficult to see how these changes affect the economy without looking at it plotted together with the inflation rate, which will therefore be analyzed below.

# Interactive relationship between the inflation and interest rate

```python
model.plot_inflation_interest(model.df)
```

The above interactive graph shows the yearly percentage change in the inflation rate and the interest rate for the United States. It can be seen that these two graphs tend to move in the same direction. However, while the inflation rate and interest rate moves in the same direction, the interest rate tends to be lagged behind the inflation rate. This is due to the fact that interest rates are the primary tool used by the United States central bank to manage inflation. A great example of this is happening right now. In 2022 when the inflation rate was skyrocketing, following the economic implications of the COVID pandemic, the interest rate started increasing rapidly, as the United States were trying to get it under control. Since 2012, the Federal Reserve have tried to maintain an annual inflation rate of 2% to maintain consistant prices, which is what we are seeing them doing right now.
